```python
import math
import time
import numpy as np
import torch
import matplotlib.pyplot as plt
```

```python
n = 10000
a = torch.ones(n)
b = torch.ones(n)
```

```python
c = torch.zeros(n)
tic = time.time()
for i in range(n):
    c[i] = a[i] + b[i]
toc = time.time()
f'{toc - tic:.5f} sec'
```

```python
def normal(x, mu, sigma):
    c = 1 / (math.sqrt(2 * math.pi * sigma **2))
    return c * np.exp( - 0.5 * (x - mu) ** 2 / sigma ** 2)
```

```python
params = [(0,1), (0,2), (3,1)]
x = np.arange(-5,5,0.01)

plt.figure(figsize=[8,3], dpi=300)
for mu, sigma in params:
      y = normal(x, mu, sigma)
      plt.plot(x, y,'-.')
plt.legend(["mean 0, std 1", "mean 0, std 2", "mean 3, std 1"])
```
