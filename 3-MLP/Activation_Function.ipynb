```python
import torch
import matplotlib.pyplot as plt
```

```python
x = torch.arange(-8.,8., 0.1,requires_grad=True)
y = torch.relu(x)
plt.figure(dpi=300)
plt.plot(x.detach(), y.detach())
plt.xlabel("x")
plt.ylabel("relu(x)")
```

```python
y.backward(torch.ones_like(x), retain_graph=True)
plt.figure(dpi=300)
plt.plot(x.detach(), x.grad)
```

```python
y = torch.sigmoid(x)
plt.figure(dpi=300)
plt.plot(x.detach(),y.detach())
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(x), retain_graph=True)
plt.figure(dpi=300)
plt.plot(x.detach(), x.grad)
```
